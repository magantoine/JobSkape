```python
def complete_approx_missing(skills, tax):
    missing = [x for x in skills if x not in tax.name.values and x != "UNK"]
    missing_esco_entries = esco[esco.preferredLabel.isin(missing)]
    
    assert(len(missing_esco_entries.index) == len(missing)) ## assert they all have exactly one entries

    merged = missing_esco_entries.merge(relationskill, on="conceptUri", how="left")\
            .merge(relationskill, left_on="broaderUri", right_on="conceptUri")\
            .drop_duplicates("preferredLabel")
            
    print(len(merged))
    merged = merged.drop(["conceptUri_y", "conceptType_y", "broaderUri_x", "broaderType_x"], axis=1).\
        rename({"conceptUri_x" : "conceptUri", "conceptType_x": "conceptType", "broaderUri_y": "broaderUri", "broaderType_y": "broaderType"}, axis=1).\
        drop_duplicates("preferredLabel")\
        [['conceptUri', 'preferredLabel', 'altLabels', 'description', 'broaderUri']]
    merged = pd.merge(merged, skillshierarchy, left_on='broaderUri', right_on='Level 3 URI', how='left')\
    [['conceptUri', 'preferredLabel', 'altLabels', 'description', 'Level 0 preferred term', 'Level 1 preferred term', 'Level 2 preferred term', 'Level 3 preferred term']]
    
    merged = merged.rename(
        columns={
            'preferredLabel': 'Type Level 4',
            'description': 'Definition',
            'Level 1 preferred term': 'Type Level 1',
            'Level 2 preferred term': 'Type Level 2',
            'Level 3 preferred term': 'Type Level 3',
            'conceptUri': 'Source',
            'Level 0 preferred term': 'Dimension'
            }
    )
    merged["unique_id"] = [tax.unique_id.max() + 1 + i for i in range(len(merged))]
    merged["name"] = merged["Type Level 4"]
    merged["name+definition"] = merged["name"] + " : " + merged["Definition"].apply(lambda x : x.lower())
    return merged
        


dev_sp_tax_complete = pd.concat([dev_sp_tax, complete_approx_missing(dev_skills_sp, dev_sp_tax)])
test_sp_tax_complete = pd.concat([test_sp_tax, complete_approx_missing(test_skills_sp, test_sp_tax)])

print(len(test_sp_tax_complete.index), len(test_skills_sp))
print(len(dev_sp_tax_complete.index), len(dev_skills_sp))
print(len(test_sp_tax_complete[test_sp_tax_complete.duplicated()].index))
print(len(dev_sp_tax_complete[dev_sp_tax_complete.duplicated()].index))
```

We already have embeddings so we'll use them and compute the rest :

```python
from transformers import AutoTokenizer, AutoModel
from tqdm.notebook import tqdm
import pickle

tqdm.pandas()


word_emb = "jjzha/jobbert-base-cased"
word_emb_tokenizer = AutoTokenizer.from_pretrained(word_emb)
word_emb_model = AutoModel.from_pretrained(word_emb)

def embedd_tax(tax):
    tax["embeddings"] = tax["name+definition"]\
                    .progress_apply(lambda nd :word_emb_model(**word_emb_tokenizer(nd, return_tensors="pt", max_length=512, padding=True, truncation=True))\
                    .last_hidden_state[:, 0, :])
    return tax
    

dev_sp_tax_complete_with_embeddings = embedd_tax(dev_sp_tax_complete)
test_sp_tax_complete_with_embeddings = embedd_tax(test_sp_tax_complete)
```
